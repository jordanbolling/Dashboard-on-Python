# External Analysis HeatMap

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Define sub-elements for each force, with associated threat levels
forces = {
    "Threat of New Entrants": [
        ("Capital Requirements", "low"),
        ("Brand Loyalty", "low"),
        ("Distribution Channels Access", "medium"),
        ('Government Policies', 'medium'),
                ('Product Placement Fairness', 'low')
    ],
    "BP of Suppliers": [
        ("No. of Key Suppliers", "medium"),
        ("Company Switching Costs", "medium"),
        ("Dependence of Supplier", "medium"),
    ],
    "BP of Buyers": [
        ("No. of Buyers v Sellers", "low"),
        ("Switching Costs", "high"),
        ("Availability of Substitutes", "medium"),
        ('Information to Switch','high')
    ],
    "Threat of Substitutes": [
        ("Variety of Alternatives", "high"),
        ("Substitute Functionality", "high"),
        ("Incumbent Brand Equity", "low")
    ],
    "Industry Rivalry": [
        ("Numerous Competitors", "high"),
        ("Industry Growth", "medium"), # yes but coke can deal
        ("Buyer Switching Costs", "medium"),
        ('Product Differentiation','high')
    ]
}

# Color codes based on threat level
threat_colors = {
    "low": "#d4edda",     # light green
    "medium": "#fff3cd",  # light yellow
    "high": "#f8d7da"     # light red
}

# Layout settings
num_forces = len(forces)
max_elements = max(len(items) for items in forces.values())
box_width = 2
box_height = 1
padding = 0.5

# Create the plot
fig, ax = plt.subplots(figsize=(box_width * num_forces + 2, box_height * max_elements + 3))
ax.set_xlim(0, num_forces * (box_width + padding))
ax.set_ylim(0, (max_elements + 2) * box_height)
ax.axis('off')

# Plot each force and its sub-elements
for i, (force, elements) in enumerate(forces.items()):
    x = i * (box_width + padding)
    
    # Draw force header (gray background)
    ax.add_patch(Rectangle((x, box_height * (max_elements + 1)), box_width, box_height,
                           facecolor="#d3d3d3", edgecolor="black"))
    ax.text(x + box_width / 2, box_height * (max_elements + 1.5), force,
            ha='center', va='center', fontsize=10, weight='bold', wrap=True)

    # Draw each element
    for j, (label, threat) in enumerate(elements):
        y = (max_elements - j) * box_height
        ax.add_patch(Rectangle((x, y), box_width, box_height,
                               facecolor=threat_colors[threat], edgecolor="black"))
        ax.text(x + box_width / 2, y + box_height / 2, label,
                ha='center', va='center', fontsize=9, wrap=True)

# Add title
plt.title("'A Tough Bottle to Open': A Profitability Analysis on the Non-Alcoholic Beverage and Soft Drink Industry",
          fontsize=13, weight='bold', pad=20)

plt.tight_layout()
plt.show()
```
